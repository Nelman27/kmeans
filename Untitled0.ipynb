## Consolidar y Preparar Datos para K-Means

### Subtask:
Crear un único DataFrame combinando `df_normal` y `df_failure`, seleccionando las columnas numéricas más significativas (unión de `top_10_failure_cols` y `top_10_normal_cols`). Añadir una columna `is_failure` para etiquetar el origen de cada fila (0 para normal, 1 para fallo). Rellenar valores nulos (si los hubiera) con 0 y luego escalar las características numéricas utilizando `RobustScaler`.


**Reasoning**:
First, I will combine the lists of top columns from both `df_failure` and `df_normal` to get a unique set of significant columns. This will ensure that all relevant features are considered for the combined DataFrame. I'll print the resulting list to confirm the selected columns.



```python
unique_significant_cols = list(set(top_10_failure_cols + top_10_normal_cols))

print(f"Columnas significativas únicas para la combinación: {unique_significant_cols}")
```

**Reasoning**:
Now that the unique significant columns have been identified, the next step is to select these columns from `df_failure` and `df_normal`, add the `is_failure` column to each, and then concatenate them into a single DataFrame `df_combined`.



```python
df_failure_selected = df_failure.select(unique_significant_cols).with_columns(pl.lit(1).alias('is_failure'))
df_normal_selected = df_normal.select(unique_significant_cols).with_columns(pl.lit(0).alias('is_failure'))

df_combined = pl.concat([df_failure_selected, df_normal_selected], how='vertical')

print(f"Dimensiones de df_combined: {df_combined.shape}")
print("Primeras 5 filas de df_combined:")
print(df_combined.head())
```

**Reasoning**:
Following the subtask instructions, I will now fill any null values in `df_combined` with 0 to ensure data consistency before scaling.



```python
df_combined = df_combined.fill_null(0)

print("Valores nulos en df_combined después de rellenar:")
print(df_combined.null_count())
```

## Entrenar Modelo K-Means y Analizar Clusters

### Subtask:
Entrenar el modelo K-Means con el valor de K óptimo determinado. Asignar las etiquetas de cluster a cada fila del DataFrame combinado y analizar la composición de cada cluster en relación con la etiqueta `is_failure` original para identificar clusters anómalos. Se generará una visualización profesional de los clusters, reduciendo la dimensionalidad si es necesario (ej. PCA 2D).


### Determinación de K Óptima

Analizando los gráficos generados en el paso anterior:

*   **Método del Codo (Elbow Method)**: Observamos que la inercia disminuye significativamente hasta K=3 o K=4, y a partir de ahí la disminución es menos pronunciada. Esto sugiere que un valor de K entre 3 y 4 podría ser un buen punto de inflexión.
*   **Coeficiente de Silueta (Silhouette Score)**: El coeficiente de silueta muestra un pico en K=2, luego una caída para K=3 y K=4 y luego un aumento para k=5 y k=6 para luego estabilizarse.

Considerando ambos métodos, un valor de `K=2` o `K=3` parece ser un buen compromiso, donde K=2 tiene el score más alto. Para este análisis, elegiremos `K=2` para empezar, ya que tiene el mejor coeficiente de silueta y el método del codo aún muestra una fuerte disminución de la inercia para K=2. Esto podría permitirnos separar de manera efectiva los casos de `failure` de los `normal`.


**Reasoning**:
Based on the visual analysis from the previous step, I have determined the optimal number of clusters (K) for the K-Means model. I will now explicitly set this value in a variable `optimal_k` and then train the `MiniBatchKMeans` model using this `optimal_k` and the scaled data `X`.



```python
optimal_k = 2

print(f"Valor de K óptimo seleccionado: {optimal_k}")

kmeans_model = MiniBatchKMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
kmeans_model.fit(X)

print("✅ Modelo MiniBatchKMeans entrenado con el K óptimo.")
```

**Reasoning**:
The K-Means model has been trained. Now, I need to assign the cluster labels obtained from the model to each row in the `df_combined` DataFrame. This will allow for further analysis of the clusters in conjunction with the original data.



```python
cluster_labels_series = pl.Series(name='cluster_label', values=kmeans_model.labels_).cast(pl.Int32)
df_combined = df_combined.with_columns(cluster_labels_series)

print("Primeras 5 filas de df_combined con las etiquetas de cluster:")
print(df_combined.head())
print("✅ Etiquetas de cluster asignadas a df_combined.")
```

**Reasoning**:
To understand the composition of each cluster, I will group the `df_combined` DataFrame by the newly added `cluster_label` and then calculate the mean of the `is_failure` column for each cluster. This will reveal the proportion of failures within each cluster, helping to identify anomalous clusters.



```python
cluster_composition = df_combined.group_by(pl.col('cluster_label').cast(pl.Int32)).agg(pl.col('is_failure').mean().alias('failure_proportion'))

print("Composición de los clusters (proporción de fallos en cada cluster):")
print(cluster_composition)
print("✅ Análisis de composición de clusters realizado.")
```

**Reasoning**:
Now that the cluster composition has been analyzed, I need to reduce the dimensionality of the data using PCA to 2 components. This will allow for a 2D visualization of the clusters. I will import `PCA` from `sklearn.decomposition` and apply it to the scaled data `X`.



```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

print(f"Dimensiones de los datos después de PCA: {X_pca.shape}")
print("✅ Dimensionalidad reducida con PCA.")
```

**Reasoning**:
To prepare the data for visualization, I will create a new DataFrame by converting `X_pca` (which contains the 2 principal components) into a Polars DataFrame, and then add the `cluster_label` and `is_failure` columns from `df_combined` to it. This will create a single DataFrame suitable for plotting.



```python
df_pca_viz = pl.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'cluster_label': df_combined['cluster_label'],
    'is_failure': df_combined['is_failure']
})

print("Primeras 5 filas de df_pca_viz:")
print(df_pca_viz.head())
print("✅ Datos preparados para visualización de clusters.")
```

**Reasoning**:
With the data prepared in `df_pca_viz`, I will now use `seaborn.scatterplot` to visualize the clusters in a 2D PCA space, coloring points by their `cluster_label` and using `is_failure` to differentiate marker styles, as per the subtask instruction.



```python
plt.figure(figsize=(15, 10)) # Aumentar el tamaño de la figura
sns.scatterplot(
    data=df_pca_viz.to_pandas(), # Convert to pandas for seaborn plotting
    x='PC1',
    y='PC2',
    hue='cluster_label',
    style='is_failure',
    palette='tab10', # Usar una paleta de colores más distintiva para categorías discretas
    s=30, # Reducir aún más el tamaño de los puntos para menor superposición
    alpha=0.5 # Aumentar la transparencia
)

# Plotting the cluster centroids
centroids_pca = pca.transform(kmeans_model.cluster_centers_)
plt.scatter(
    centroids_pca[:, 0],
    centroids_pca[:, 1],
    marker='X', # Marcador distintivo para centroides
    s=700, # Tamaño grande para centroides
    color='black',
    label='Centroides de Cluster',
    edgecolors='white', # Borde blanco para visibilidad
    linewidth=2
)

# Añadir etiquetas de proporción de fallos a los centroides
for i, centroid in enumerate(centroids_pca):
    # Obtener la proporción de fallos para el cluster actual, asegurando la comparación de tipos
    failure_prop = cluster_composition.filter(pl.col('cluster_label').cast(pl.Int32) == i).select('failure_proportion').item()
    plt.text(centroid[0], centroid[1] + 0.5, # Ajustar la posición del texto
             f'K{i}: {failure_prop:.2%}',
             horizontalalignment='center',
             fontsize=12,
             color='red', # Color rojo para destacar la proporción de fallos
             weight='bold')

plt.title('Visualización de Clusters K-Means (PCA 2D) con Centroides y Proporción de Fallos', fontsize=18, pad=20)
plt.xlabel('Componente Principal 1', fontsize=14)
plt.ylabel('Componente Principal 2', fontsize=14)
plt.legend(title='Etiqueta / Fallo', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0, 0.98, 0.98]) # Ajustar layout para evitar cortes de leyendas
plt.show()

print("✅ Visualización de clusters con centroides y proporciones de fallo generada y mejorada.")
```
